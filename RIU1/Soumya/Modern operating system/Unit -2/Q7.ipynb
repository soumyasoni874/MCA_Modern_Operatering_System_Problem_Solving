```python
def round_robin(processes, burst_time, time_quantum):
    n = len(processes)

    remaining_bt = burst_time[:]   # Copy of burst times
    waiting_time = [0] * n
    turnaround_time = [0] * n
    t = 0  # Current time
    gantt = []  # To store Gantt chart sequence

    # Continue until all burst times become zero
    while True:
        done = True
        for i in range(n):
            if remaining_bt[i] > 0:
                done = False
                gantt.append(processes[i])

                if remaining_bt[i] > time_quantum:
                    t += time_quantum
                    remaining_bt[i] -= time_quantum
                else:
                    t += remaining_bt[i]
                    waiting_time[i] = t - burst_time[i]
                    remaining_bt[i] = 0

        if done:
            break

    # Calculate turnaround time
    for i in range(n):
        turnaround_time[i] = waiting_time[i] + burst_time[i]

    # Final output
    print("\nRound Robin Scheduling (Time Quantum =", time_quantum, ")")
    print("Process\tBurst Time\tWaiting Time\tTurnaround Time")
    for i in range(n):
        print(f"{processes[i]}\t   {burst_time[i]}\t\t   {waiting_time[i]}\t\t      {turnaround_time[i]}")

    # Gantt Chart
    print("\nGantt Chart:")
    for p in gantt:
        print(f"| {p} ", end="")
    print("|")



# Example
processes = ["P1", "P2", "P3", "P4"]
burst_time = [10, 4, 5, 6]
time_quantum = 3

round_robin(processes, burst_time, time_quantum)
```
